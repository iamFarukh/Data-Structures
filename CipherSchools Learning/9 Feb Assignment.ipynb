## Problem Statement
**Generate the inOrder and postOrder using given PreOrder**


```python
# Binary Search Tree

class Node:
    def __init__(self,key):
        self.key = key
        self.left = None
        self.right = None
        
        
def inOrder(node):
    if(node):
        inOrder(node.left)
        print(node.key, end = " ")
        inOrder(node.right)
        
def preOrder(node):
    if(node):
        print(node.key, end = " ")
        inOrder(node.left)
        inOrder(node.right)
        
def postOrder(node):
    if(node):
        inOrder(node.left)
        inOrder(node.right)
        print(node.key, end = " ")
        
def insert(root,key):
    #Base Condition
    if root is None:
        return Node(key)
    
    else:
        # key < root.data | key > root.data
        if key < root.key:
            root.left = insert(root.left,key)
        elif key > root.key:
            root.right = insert(root.right,key)
        else:
            return root
    return root
            

#Driver Method

root = Node(10)
root = insert(root,22)
root = insert(root,7)
root = insert(root,20)
root = insert(root,70)
root = insert(root,22)
root = insert(root,30)

inOrder(root)
print()
                                                                             
    
        
```

## Problem Statement
**Generate the inOrder and postOrder using given PreOrder**


```python
# Take Pre-Order as Input from User 
# number of elements in Tree
n = int(input("Enter number of elements : ")) 
  
# Below line read inputs from user using map() function  
a = list(map(int,input("\nEnter the elements : ").strip().split()))[:n] 

# Creating BST with Given Pre-Order
# in Pre-Order the First Element always the Root 
root = Node(a[0])

# Rest element is Inserted as the Properties of BST
for i in range(1,n):
    root = insert(root,a[i])
    
#Displaying The In-Order
print("Displaying The In-Order : ",end =" ")
inOrder(root)
print()

#Displaying The Pre-Order
print("Displaying The Pre-Order : ",end =" ")
preOrder(root)
print()

#Displaying The Post-Order
print("Displaying The Post-Order : ",end =" ")
postOrder(root)
print()
    
```

## Problem Statement
**Generate the inOrder and preOrder using given PostOrder**

```python
# Take Pre-Order as Input from User 
# number of elements in Tree
n = int(input("Enter number of elements : ")) 
  
# Below line read inputs from user using map() function  
a = list(map(int,input("\nEnter the elements : ").strip().split()))[:n] 

# Creating BST with Given Pre-Order
# in Post-Order the last Element always the Root 
root = Node(a[n-1])

# Rest element is Inserted as the Properties of BST
for i in range(0,n-1):
    root = insert(root,a[i])
    
#Displaying The In-Order
print("Displaying The In-Order : ",end =" ")
inOrder(root)
print()

#Displaying The Pre-Order
print("Displaying The Pre-Order : ",end =" ")
preOrder(root)
print()

#Displaying The Post-Order
print("Displaying The Post-Order : ",end =" ")
postOrder(root)
print()
    
```

```python
class Node:
    # constructor
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    # constructor of LL
    def __init__(self):
        self.head = None
        
    # ADD METHODS
    
    # 1. Add at the begining of the LL
    def addBegining(self, data):
        newNode = Node(data)
        newNode.next = self.head
        self.head = newNode
        
    # 2. Add at the end of the LL
    def addEnd(self, data):
        # create the Node, paste the data
        newNode = Node(data)
        
        # check if the LL is empty
        if self.head is None:
            # this Node will be the I node
            self.head = newNode
            return
        
        # LL already exists
        # traverse to the end, and change the next of the last node
        temp = self.head
        #traverse the LL
        while(temp.next is not None):
            temp = temp.next
        temp.next = newNode
        
    # 3. Add at the middle of a given node in the LL
    def addMid(self, midNode, data):
        # condition -> midNode == Node
        if(midNode == None):
            print("Node is NOT available")
            return
        
        # create the node, and do the linking
        newNode = Node(data)
        newNode.next = midNode.next
        midNode.next = newNode
        
    # DELETE METHOD
    
    def remove(self, data):
        temp = self.head
        
        # LL does not exist
        if(temp is not None):
            # case 1: remove the first Node
            if(self.head.data == data):
                self.head = temp.next
                temp = None
                return
            
        while(temp is not None):
            # search for the data
            if(temp.data == data):
                break
            prev = temp
            temp = temp.next
            
        # if the data is not found
        if(temp == None):
            print(data, "does not exist in the LinkedList")
            return
            
        # exchange the links of prev.next and node.next
        prev.next = temp.next
        temp = None
            
            
    def display(self):
        temp = self.head
        print("The LinkedList is: ")
        while(temp):
            print(temp.data, end = " ")
            temp = temp.next
        print()
        
    def find_middle(self):
        count = 0
        temp = self.head
        while(temp):
            count += 1
            temp = temp.next
        mid = count//2
        temp = self.head
        count = 0
        while(count != mid):
            temp = temp.next
            count += 1
        print(print(temp.data)
            
            
        

            
        
# DRIVER CODE

linked_list = LinkedList()
linked_list.addEnd(10)
linked_list.addEnd(20)
linked_list.addEnd(30)
linked_list.addEnd(40)
linked_list.addEnd(50)
linked_list.addEnd(60)
linked_list.display()
linked_list.find_middle()
```
